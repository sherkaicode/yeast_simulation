## Description of the yeast
The yeast shall have a dna strand of "xx xx xx xx xx xx"
which has the probability of mutation every reproduction (the offspring) and during presence of radiation (environment)

## Properties of the Yeast
1. fitness score = defined by the function of the genes (FA, SS): with (0.3, 0.7) weights [0, 1]
2. reproduction rate = normal (90 mins), slow (120 mins), fast (60 mins)
3. lifespan = normal (6 hours), high (9 hours), low (3 hour), very low (1 hour)
4. fitness ability = normal (60), high (90), low (30), very low (10)
5. food intake = normal (1), high (2), low (0.5), very low (0)
6. Survival Stimuli = normal (60, high (90), low (30), very low (10)
7. Innate energy = normal (100), high (150), low (70), very low (50)
8. Byproducts = [carbon dioxide, water, alcohol]

## Properties of the Environment
- oxygen count
- sugar count

```python
import numpy as np
import random
```

```python
#Dictionary for dna
class Yeast:
    def __init__(self, dna, offsprings = 0, age = 0):
        #Dictionary for dna
        prio_dict = {"aa": 3, "ab": 2, "ac": 1, "bb": 3, "ba": 2, "bc": 2, "ca": 1, "cb": 1, "cc": 0}
        
                
        self.dna = dna
        self.dna_vals = []
        self.offsprings = offsprings
        self.age = age
        self.byproducts = []
        
        

        #Values for fitness 0
        dict_fa = [10, 30, 60, 90]
        #Values for food intake 1
        dict_fi = [0, 0.5, 1, 2]
        #Values for lifespan 2
        dict_ls = [60, 3*60, 6*60, 9*60]
        #Values for UP 3
        dict_up = [1, 1, 1, 1]
        #Values for Survival stimuli 4
        dict_ss = [10, 30, 60, 90]
        #Values for innate energy 5
        dict_ie = [50, 70, 100, 150]

        gene_array = [dict_fa, dict_fi, dict_ls, dict_up, dict_ss, dict_ie]
        for i, gene in enumerate(dna):
            self.dna_vals.append(gene_array[i][prio_dict[gene]])
        
        
    def respiration(self, env_sugar, env_oxygen, env_co2, env_water, env_alcohol):
        if env_sugar == 0:
            self.dna_vals[5] -= 10
        else:    
            if env_oxygen != 0:
                #Consume
                env_sugar -= 1
                env_oxygen -= 1
                #gain
                self.dna_vals[5] += 15
                #Byproducts
                env_co2 += 1
                env_water += 1
            else:
                #Consume
                env_sugar -= 1
                #Gain
                self.dna_vals[5] += 5
                #Byproducts
                env_co2 += 1
                env_alcohol += 1
        
    def grow(self, env_sugar, env_oxygen, env_co2, env_water, env_alcohol):
        self.age += 1
        self.respiration(env_sugar, env_oxygen, env_co2, env_water, env_alcohol)

    def die(self, yeast_population_tracker, time):
        if (self.age == self.dna_vals[2]) or (self.dna_vals[5] <= 0):
            yeast_population_tracker.remove(self)
        
        
    def show(self):
        print("My DNA: ", self.dna)
        print("My total offsprings:", self.offsprings)
        print("My age: ", self.age, "minutes")

#Properties of the environment
sugar_count = 100
oxygen_count = 100
co2_count = 100
water_count = 100
alcohol_count = 0
yeast_population_tracker = []
#main simulation 
initial_number_of_yeast = 10
for i in range(initial_number_of_yeast):
    yeast_population_tracker.append(Yeast(["".join(random.choices("abc", k=2)) for i in range(6)]))

#Add test yeast
yeast_population_tracker.append(Yeast(["aa", "aa", "cc", "aa", "aa", "aa"]))
for time in range(3*60):

    #Update each Yeast
    for yeast in yeast_population_tracker:
        yeast.grow(sugar_count, oxygen_count, co2_count, water_count, alcohol_count)
        yeast.die(yeast_population_tracker, time)

    #6 Hours specific check up
    # if i%(60) == 0:
    #     for yeast in yeast_population_tracker:
    #         yeast.show()

    if time%60 == 0:
        print(len(yeast_population_tracker))
            
    
```
