```python
import numpy as np
import random
import uuid
```

```python
#Dictionary for dna
class Yeast:
    def __init__(self, y_id, dna, size, gen, parent, kid = None,offsprings = 0, age = 0):
        #Dictionary for dna
        prio_dict = {"aa": 3, "ab": 2, "ac": 1, "bb": 3, "ba": 2, "bc": 2, "ca": 1, "cb": 1, "cc": 0}

        self.parent = parent
        self.kid = kid
        self.id = y_id
        self.dna = dna
        self.dna_vals = []
        self.offsprings = offsprings
        self.age = age
        self.size = size
        self.gen = gen
        self.byproducts = []
        
        

        #Values for fitness 0
        dict_fa = [10, 30, 60, 90]
        #Values for food intake 1
        dict_fi = [0, 0.5, 1, 2]
        #Values for lifespan 2
        dict_ls = [60, 3*60, 6*60, 9*60]
        #Values for UP 3
        dict_up = [1, 1, 1, 1]
        #Values for Survival stimuli 4
        dict_ss = [10, 30, 60, 90]
        #Values for innate energy 5
        dict_ie = [50, 70, 100, 150]

        gene_array = [dict_fa, dict_fi, dict_ls, dict_up, dict_ss, dict_ie]
        for i, gene in enumerate(dna):
            self.dna_vals.append(gene_array[i][prio_dict[gene]])
        
        
    # def respiration(self, env_sugar, env_oxygen, env_co2, env_water, env_alcohol):
    #     if env_sugar == 0:
    #         self.dna_vals[5] -= 10
    #     else:    
    #         if env_oxygen != 0:
    #             #Consume
    #             env_sugar -= self.dna_vals[1]
    #             env_oxygen -= 1
    #             #gain (energy, fitness, size)
    #             self.dna_vals[5] += 15
    #             self.dna_vals[0] += 5
    #             self.size += 4
    #             #Byproducts
    #             env_co2 += 1
    #             env_water += 1
    #         else:
    #             #Consume
    #             env_sugar -= self.dna_vals[1]
    #             #Gain (energy, fitness, size)
    #             self.dna_vals[5] += 5
    #             self.dna_vals[0] += 1
    #             self.size += 2
    #             #Byproducts
    #             env_co2 += 1
    #             env_alcohol += 1
    def respiration(self, env_prop):
        if env_prop[0] <= 0 or env_prop[0] < self.dna_vals[1]:
            self.dna_vals[5] -= 10
        else:    
            if env_prop[1] != 0:
                #Consume
                env_prop[0] -= self.dna_vals[1]
                env_prop[1] -= 1
                #gain (energy, fitness, size)
                self.dna_vals[5] += 15
                self.dna_vals[0] += 5
                self.size += 4
                #Byproducts (CO2, water)
                env_prop[2] += 1
                env_prop[3] += 1
            else:
                #Consume
                env_prop[0] -= self.dna_vals[1]
                #Gain (energy, fitness, size)
                self.dna_vals[5] += 5
                self.dna_vals[0] += 1
                self.size += 2
                #Byproducts (CO2, alcohol)
                env_prop[2] += 1
                env_prop[4] += 1
        
    # def grow(self, env_sugar, env_oxygen, env_co2, env_water, env_alcohol):
    #     self.age += 1
    #     self.respiration(env_sugar, env_oxygen, env_co2, env_water, env_alcohol)

    def grow(self, env_prop):
        self.age += 1
        self.respiration(env_prop)

    def die(self, yeast_population_tracker, time):
        if (self.age == self.dna_vals[2]) or (self.dna_vals[5] <= 0):
            yeast_population_tracker.remove(self)
    
    def reproduce(self, yeast_population_tracker):
        if self.age%90 == 0:     
            if (self.size >= 20) and (self.dna_vals[5] >= 50) and (self.dna_vals[0] >= 30):
                self.size = self.size//2
                self.dna_vals[5] = self.dna_vals[5]//2
                self.dna_vals[0] -= 5
                
                yeast_population_tracker.append(Yeast( uuid.uuid4(),self.dna, self.size//2, self.gen + 1, self))
                self.kid = yeast_population_tracker[-1]
                self.offsprings += 1
    def show(self):
        print("My DNA: ", self.dna)
        print("My total offsprings:", self.offsprings)
        print("My age: ", self.age, "minutes")

#Properties of the environment
sugar_count = 10000 #0
oxygen_count = 100 #1
co2_count = 100 #2
water_count = 100 #3
alcohol_count = 0 #4
properties_array = [sugar_count, oxygen_count, co2_count, water_count, alcohol_count]
yeast_population_tracker = []
#main simulation 
initial_number_of_yeast = 10
for i in range(initial_number_of_yeast):
    yeast_population_tracker.append(Yeast(uuid.uuid4(),["".join(random.choices("abc", k=2)) for i in range(6)], 2*random.randint(1, 3), 0, None))

#Add test yeast
yeast_population_tracker.append(Yeast(uuid.uuid4(),["aa", "aa", "aa", "aa", "aa", "aa"], 20, 0, None))

for time in range(10*60):
    #Update each Yeast
    for yeast in yeast_population_tracker:
        # yeast.grow(sugar_count, oxygen_count, co2_count, water_count, alcohol_count)
        yeast.grow(properties_array)
        yeast.die(yeast_population_tracker, time)
        yeast.reproduce(yeast_population_tracker)

        #Numpy Array Tracker

    if time%60 == 0:
        print("Number of Yeast:", len(yeast_population_tracker))
        print("Properties array:", properties_array)
            
    
```
